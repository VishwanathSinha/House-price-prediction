```python
uv=np.percentile(df.n_hot_rooms,[99])[0]
```

The correlation between air quality and parks is 0.91.

This means that these two independent variables are highly correlated.

And if we take both of these variables for our analysis, this may lead to multicollinearity.

So we have to delete one of these two variables.

There are multiple ways to select which variable to delete.

First one is to check the correlation of these two variables with our dependent y variable.

So, for example, For park, we have a correlation of -0.39 and for air quality We have a correlation of 0.42.

We will take the variable, which has higher correlation with our dependent variable.

So in this case, this is air quality.

You can also use your business, knowledge and business underdtanding

To select one of these two variables.

So if some variable is easy to measure and is easily understandable.

take that variable and delete

Another variable.

For our this example, we are going to delete park from our dataset.

```python
# ridge and lasso regression
```

```python
from sklearn import preprocessing
#standardized our independent variables
```

```python
scaler=preprocessing.StandardScaler().fit(x_train)
```

```python
x_train_s=scaler.transform(x_train)
```

```python
x_test_s=scaler.transform(x_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
lm_r=Ridge(alpha=0.5)
lm_r.fit(x_train_s,y_train)
```

```python
r2_score(y_test,lm_r.predict(x_test_s))
```

```python
#we can also change the value of alpha while running ridge regression and we want to 
#find value of alpha for which r2 score is maximum
#for trying out values of hyperparameters we use validation curves
```

```python
from sklearn.model_selection import validation_curve
```

```python
validation_curve?
```

```python
param_range=np.logspace(-2,8,100)
#create 100 values between 10^-2 to 10^8 
param_range
```

```python
#iterations for our model
train_scores,test_scores=validation_curve(Ridge(),x_train_s,y_train,"alpha",param_range,scoring='r2')
```

```python
#this is basically running a k-fold validation on our test data 
print(train_scores)
print(test_scores)
```

```python
train_mean=np.mean(train_scores,axis=1)
```

```python
test_mean=np.mean(test_scores,axis=1)
```

```python
train_mean
```

```python
#we want highest r2 value
max(test_mean)
```

```python
sns.jointplot(x=np.log(param_range),y=test_mean)
```

```python
#to find the alpha for max r2 score
np.where(test_mean==max(test_mean))
```

```python
param_range[31]
#required alpha value
```

```python
lm_r_best=Ridge(alpha=param_range[31])
```

```python
lm_r_best.fit(x_train_s,y_train)
```

```python
r2_score(y_test,lm_r_best.predict(x_test_s))
```

```python
r2_score(y_train,lm_r_best.predict(x_train_s))
```
